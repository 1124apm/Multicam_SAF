```python
import os
import time
import google.generativeai as genai

# ---------------------------------------------------------
# 1. API ì„¤ì •
# ---------------------------------------------------------
# âš ï¸ ì£¼ì˜: ê¸°ì¡´ í‚¤ê°€ ìœ ì¶œë˜ì—ˆìœ¼ë¯€ë¡œ ì¬ë°œê¸‰ ë°›ì€ ìƒˆ í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”.
API_KEY = "ì—¬ê¸°ì—_ìƒˆë¡œìš´_API_KEYë¥¼_ì…ë ¥í•˜ì„¸ìš”" 
genai.configure(api_key=API_KEY)

# ëª¨ë¸ ì„¤ì •
model = genai.GenerativeModel('gemini-flash-latest')

# ---------------------------------------------------------
# 2. íŒŒì¼ ì½ê¸° í•¨ìˆ˜ (ë‹¨ì¼ íŒŒì¼)
# ---------------------------------------------------------
def read_text_from_file(file_path):
    """ì§€ì •ëœ ê²½ë¡œì˜ íŒŒì¼ ë‚´ìš©ì„ ì½ì–´ì„œ ë¬¸ìì—´ë¡œ ë°˜í™˜"""
    if not os.path.exists(file_path):
        print(f"âš ï¸ íŒŒì¼ ì—†ìŒ: {file_path}")
        return None
    
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            return f.read()
    except Exception as e:
        print(f"âš ï¸ íŒŒì¼ ì½ê¸° ì˜¤ë¥˜ ({file_path}): {e}")
        return None

# ---------------------------------------------------------
# 3. LLM í‚¤ì›Œë“œ ì¶”ì¶œ í•¨ìˆ˜
# ---------------------------------------------------------
def extract_vibe_with_llm(team_name, text_data):
    if not text_data:
        print(f"âŒ {team_name}: ë¶„ì„í•  ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
        return "ë°ì´í„° ì—†ìŒ"

    # ì…ë ¥ ê¸¸ì´ ì œí•œ (ë¹„ìš© ë° ì¿¼í„° ê´€ë¦¬)
    input_text = text_data[:20000] # ë¶„ì„ì— ì¶©ë¶„í•˜ë„ë¡ 2ë§Œ ì ì •ë„ ì‚¬ìš©
    
    print(f"â–¶ [{team_name}] ë¶„ì„ ìš”ì²­ ì¤‘... (ì „ì†¡ ë°ì´í„°: {len(input_text)}ì)")

    prompt = f"""
    ë„ˆëŠ” ìŠ¤í¬ì¸  ë°ì´í„° ë¶„ì„ ì „ë¬¸ê°€ì•¼. ì•„ë˜ í…ìŠ¤íŠ¸ëŠ” F1(í¬ë®¬ëŸ¬ 1) íŒ€ '{team_name}'ì— ëŒ€í•œ ë°ì´í„°ì•¼.
    ì´ íŒ€ì˜ **ê³ ìœ í•œ ì„±í–¥, ë ˆì´ì‹± ìŠ¤íƒ€ì¼, íŒ¬ë¤ ë¶„ìœ„ê¸°**ë¥¼ ë‚˜íƒ€ë‚´ëŠ” í•µì‹¬ í‚¤ì›Œë“œ 15ê°œë¥¼ í•œê¸€ë¡œ ë²ˆì—­í•´ì„œ ì¶”ì¶œí•´ì¤˜.
    
    [ì¡°ê±´]
    1. 'F1', 'ë ˆì´ì‹±', 'ìë™ì°¨', 'ìš°ìŠ¹', 'ë“œë¼ì´ë²„' ê°™ì€ ë»”í•œ ë‹¨ì–´ ì œì™¸.
    2. í˜•ìš©ì‚¬ë‚˜ ëª…ì‚¬ ìœ„ì£¼ (ì˜ˆ: í˜ì‹ ì ì¸, ê³µê²©ì ì¸ ì „ëµ, ê·€ì¡±ì ì¸, ì••ë„ì ì¸).
    3. ê²°ê³¼ëŠ” ì˜¤ì§ í‚¤ì›Œë“œë§Œ ì‰¼í‘œ(,)ë¡œ êµ¬ë¶„í•´ì„œ ì¶œë ¥í•´.
    
    [í…ìŠ¤íŠ¸]
    {input_text}
    """

    try:
        response = model.generate_content(prompt)
        if response.text:
            return response.text.strip()
        else:
            return "ì‘ë‹µ ì—†ìŒ"
    except Exception as e:
        print(f"âš ï¸ API í˜¸ì¶œ ì—ëŸ¬: {e}")
        return "Error"

# ---------------------------------------------------------
# 4. ì‹¤í–‰ ë°ì´í„° ì„¤ì •
# ---------------------------------------------------------
# ë”•ì…”ë„ˆë¦¬ êµ¬ì¡° ë³€ê²½: { "íŒ€ì´ë¦„": "íŒŒì¼ê²½ë¡œ" }
teams = {
    "Jeju Test": r"ê²½ë¡œ" 
}

# ---------------------------------------------------------
# 5. ë©”ì¸ ë£¨í”„ ì‹¤í–‰
# ---------------------------------------------------------
results = {}

print("ğŸš€ LLM F1 íŒ€ ì„±í–¥ ë¶„ì„ ì‹œì‘ (ë‹¨ì¼ íŒŒì¼ ëª¨ë“œ)...\n")

for team_name, file_path in teams.items():
    
    # 1. íŒŒì¼ì—ì„œ í…ìŠ¤íŠ¸ ì½ê¸°
    text_data = read_text_from_file(file_path)
    
    # 2. ë°ì´í„°ê°€ ìˆì„ ê²½ìš°ì—ë§Œ ë¶„ì„ ì§„í–‰
    if text_data:
        keywords = extract_vibe_with_llm(team_name, text_data)
        results[team_name] = keywords
        
        print(f"=== ğŸ“Š {team_name} ê²°ê³¼ ===")
        print(keywords)
        print("-" * 50 + "\n")
        
        # ì†ë„ ì¡°ì ˆ
        print("â³ API ê³¼ë¶€í•˜ ë°©ì§€ë¥¼ ìœ„í•´ 5ì´ˆ ëŒ€ê¸° ì¤‘...")
        time.sleep(5)
    else:
        print(f"pass: {team_name} (íŒŒì¼ì„ ì½ì§€ ëª»í•¨)\n")
```
